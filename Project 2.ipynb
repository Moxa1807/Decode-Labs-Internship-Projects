## Distribution of Quantity

This analysis shows how many products customers purchase in each order.

```python
plt.figure(figsize=(8,5))

sns.histplot(df["Quantity"], bins=20, kde=True)

plt.title("Distribution of Quantity")
plt.xlabel("Quantity")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(8,2))

sns.boxplot(x=df["Quantity"])

plt.title("Boxplot of Quantity")

plt.show()
```

### Insight

- The distribution of order quantity is concentrated around lower values, indicating that most customers purchase a small number of products per transaction.
- A limited number of orders contain higher quantities, suggesting occasional bulk purchases.
- The boxplot reveals a few higher-value observations, but these represent only a small portion of the dataset.

## Which products are ordered the most?

```python
top_products = df["Product"].value_counts()

top_products
```

```python
plt.figure(figsize=(10,6))

sns.countplot(
    y="Product",
    data=df,
    order=df["Product"].value_counts().index
)

plt.title("Number of Orders by Product")

plt.xlabel("Number of Orders")

plt.ylabel("Product")

plt.show()
```

### Insight

- Product demand is fairly balanced across all product categories.
- Printer recorded the highest number of customer orders, followed closely by Tablet and Chair.
- Phone received comparatively fewer orders, indicating relatively lower purchase frequency.

### This section identifies the revenue with products

```python
product_revenue = df.groupby("Product")["TotalPrice"].sum().sort_values(ascending=False)

product_revenue
```

```python
plt.figure(figsize=(10,5))

product_revenue.plot(kind="bar")

plt.title("Revenue Generated by Product")
plt.xlabel("Product")
plt.ylabel("Total Revenue")

plt.xticks(rotation=45)

plt.show()
```

### Insight

- Revenue generation differs from order frequency, indicating that product price also plays an important role.
- Chair generated the highest overall revenue despite having fewer orders than Printer.
- High-value products contribute significantly to total revenue even if their sales volume is comparatively lower.

## Monthly Sales Trend

```python
month_order = [
    "January","February","March","April","May","June",
    "July","August","September","October","November","December"
]

df["Month"] = pd.Categorical(
    df["Month"],
    categories=month_order,
    ordered=True
)
```

```python
monthly_sales = df.groupby("Month")["TotalPrice"].sum()

monthly_sales
```

```python
plt.figure(figsize=(10,5))

monthly_sales.plot(marker="o")

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")

plt.grid(True)

plt.show()
```

### Insight

- Sales fluctuate across different months instead of remaining constant.
- Certain months contribute noticeably higher revenue, indicating seasonal variations in customer purchasing behavior.
- Understanding these monthly trends can help businesses improve inventory planning and promotional campaigns during peak periods.

## Correlation Analysis

The heatmap shows the relationship between numerical variables.

```python
correlation = df[
    ["Quantity","UnitPrice","ItemsInCart","TotalPrice"]
].corr()

correlation
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    correlation,
    annot=True,
    cmap="Blues",
    fmt=".2f"
)

plt.title("Correlation Heatmap")

plt.show()
```

### Insight

- Total Price has a strong positive correlation with Unit Price and Quantity, which is expected because order value depends on both factors.
- Items in Cart shows only a moderate relationship with Total Price, suggesting that the number of products added to the cart does not always translate into higher revenue.
- No strong negative correlations are observed among the numerical variables.
